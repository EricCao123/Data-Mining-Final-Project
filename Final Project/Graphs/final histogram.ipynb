```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('before_dirty.csv', usecols=['Percentage']) #reads file to the variable data
data2 = pd.read_csv('before_clean.csv', usecols=['Percentage']) #reads file to the variable data2
```

```python
#The values come with a percentage sign. these two for loops remove the percentage sign at the end of the values
#in data and data 2
for i in range(len(data['Percentage'])): 
    data['Percentage'][i] = data['Percentage'][i][:-1]
for i in range(len(data2['Percentage'])):
    data2['Percentage'][i] = data2['Percentage'][i][:-1]
```

```python
#The csv file contains string and not floats. these for loops change the string values into float values
ytest = []
for y in data['Percentage']:
    if float(y) > 0.0:
        ytest = ytest + [float(y)]
xtest = []
for x in data2['Percentage']:
    if float(x) > 0.0:
        xtest = xtest + [float(x)]
```

```python
#Histogram for data, before_dirty.csv
plt.hist(ytest, bins=10, color=['green'], edgecolor='black')
plt.rcParams["figure.figsize"] = (20,9)
plt.title('Percentages of Target Word that was Preceded by Most Common Word')
plt.xlabel('Percentage')
plt.ylabel('No of times')
plt.rcParams.update({'font.size': 16})
plt.show()
#Histogram for data2, before_clean.csv
plt.hist(xtest, bins=10, color=['red'], edgecolor='black')
plt.rcParams["figure.figsize"] = (20,9)
plt.title('Percentages of Target Word that was Preceded by Most Common Word (Target Word Appeared Greater Than 1000 Times)')
plt.xlabel('Percentage')
plt.ylabel('No of times')
plt.rcParams.update({'font.size': 16})
plt.show()
#NOTE: If graphs appear small, run this snippet again
```
