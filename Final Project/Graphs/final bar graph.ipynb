```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('data.csv', names=['word', 'number']) #Reads the data into the variable called data
#The first 5 rows are word counts based on word type. 
#This is not needed for this program so the line below removes them from the array
data = data.iloc[6:] 
```

```python
#opens word dictionaries and writes them to respective variables
with open('all_words.txt') as word_file:
    valid_words = set(word_file.read().split())
with open('adult_words.txt') as word_file:
    adult_words = set(word_file.read().split())
with open('hate_words.txt') as word_file:
    hate_words = set(word_file.read().split())
with open('function_words.txt') as word_file:
    function_words = set(word_file.read().split())
```

```python
#this changes the array values from string to int
xarray = []
yarray = []
all_words = []
all_numbers = []
xadult = []
yadult = []
xhate = []
yhate = []
for x in data.word:
    xarray = xarray + [x]

for y in data.number:
    yarray = yarray + [int(y)]
    
for z in range(len(xarray)):
    if xarray[z] in valid_words:
        
        if xarray[z] in adult_words:
            xadult.append(xarray[z])
            yadult.append(yarray[z])
        elif xarray[z] in hate_words:
            xhate.append(xarray[z])
            yhate.append(yarray[z])
        elif xarray[z] not in function_words:
            all_words.append(xarray[z])
            all_numbers.append(yarray[z])
```

```python

# Create bar graph of 20 most common words
plt.bar(all_words[:20],all_numbers[:20],color=['black', 'red', 'green', 'blue', 'cyan'])
plt.rcParams["figure.figsize"] = (20,9)
# Add title and axis names
plt.title('Top 20 Frequently Seen Words in 100 Books')
plt.xlabel('word')
plt.ylabel('word frequency')
plt.rcParams.update({'font.size': 16})
#NOTE: If the graphs appear small, run this snippet again
```

```python
# Create bar graph of 20 most common words
plt.bar(xadult[:20],yadult[:20],color=['black', 'red', 'green', 'blue', 'cyan'])
plt.rcParams["figure.figsize"] = (20,9)
# Add title and axis names
plt.title('Top 20 Frequently Seen Adult Words in 100 Books')
plt.xlabel('word')
plt.ylabel('word frequency')
plt.rcParams.update({'font.size': 14})
```

```python
# Create bar graph of 20 most common words
plt.bar(xhate[:20],yhate[:20],color=['black', 'red', 'green', 'blue', 'cyan'])
plt.rcParams["figure.figsize"] = (20,9)
# Add title and axis names
plt.title('Top 7 Frequently Seen Hate Words in 100 Books')
plt.xlabel('word')
plt.ylabel('word frequency')
plt.rcParams.update({'font.size': 14})
```

```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd
data = pd.read_csv('hateAdultBookCount.csv', names=['type', 'number'])
```

```python
# Values of each group
bars1 = int(data['number'][1])
bars2 = int(data['number'][2])
bars3 = int(data['number'][3])

bars = np.add(bars1, bars2).tolist()
r = ['Total'] #x-tick label
plt.bar(r, bars1, color='#ff69b4', edgecolor='black') #color for normal
plt.bar(r, bars2, bottom=bars1, color='#39ff14', edgecolor='black') #color for hate
plt.bar(r, bars3, bottom=bars1, color='#1b03a3', edgecolor='black') #color for adult
plt.legend(('Normal', 'Hate', 'Adult'), prop={'size': 20})
plt.title('Comparison of Normal Books, Hate Books, and Adult Books')
plt.ylabel('Number of Books')
plt.show()
```
