```python
#change directory into /Data where all the data files are located
os.chdir(os.getcwd()+"/Data")
#acquire the current path
cwd = os.getcwd()
#acquire a list of all the current directories
directories = os.listdir()
#initialize the total amount of adult books
totalAdult = 0
#initialize the total amount of hate books
totalHate = 0
#initialize the total amount of normal books
totalBooks = 0
#initialize the list that will contain all the hate books
hateBooks = []
#initialize the list that will contain all the normal books
normalBooks = []
#initialize the list that will contain all the adult books
adultBooks = []
#iterate over all the directories that we are currently in
for directory in directories:
    #if our current directory object is a directory and is not a python notebook checpoint continue
    if os.path.isdir(directory) and directory != ".ipynb_checkpoints":
        #change into the directory
        os.chdir(os.getcwd()+"/"+directory)
        #acquire a list of all the sub directories in our current directory
        subDirectories = os.listdir()
        #open a new csv file or overwrite the existing file and proceed to write in it
        with open('hateAdultBookCount.csv', mode = 'w') as data:
            #initialize the writer object for our file
            data_writer = csv.writer(data, delimiter=',', quoting=csv.QUOTE_MINIMAL)
            #write the header line for our csv file
            data_writer.writerow(["book type","number"])
            #initialize the hate amount to 0
            hate = 0;
            #initialize the adult amount to 0
            adult = 0;
            #initialize the normal amount to 0
            total = 0;
            #iterate over all the sub directories in our current directory
            for subDirectory in subDirectories:
                #if our current directory object is a directory and is not a python notebook checpoint continue
                if os.path.isdir(subDirectory) and subDirectory != ".ipynb_checkpoints":
                    #change into our subdirectory
                    os.chdir(os.getcwd()+"/"+subDirectory)
                    #grab the files in our sub directory
                    files = os.listdir()
                    #iterate over all the files in our sub directory
                    for file in files:
                        #if the file is the book data file
                        if file != "after.csv" and file!= "before.csv" and file != "hateAdultBookCounter.csv" and not(os.path.isdir(file)):
                            #read the current csv file and assign it to our variable
                            data = pd.read_csv(file,names=['word','number'])
                            #if this book contains hate words
                            if int(data.iloc[2:3]['number']) > 0:
                                #increase our hate counter
                                hate+=1;
                                #increase our total hate counter
                                totalHate+=1;
                                #add this file to our list of hate books
                                hateBooks.append(file)
                            #if this book contains adult language
                            if int(data.iloc[3:4]['number']) > 0:
                                #increase our adult counter
                                adult+=1;
                                #increase our total adult counter
                                totalAdult+=1;
                                #add this file to our list of adult books
                                adultBooks.append(file)
                            #if this file does not contain hate words or adult language
                            if not int(data.iloc[2:3]['number']) > 0 and not int(data.iloc[3:4]['number']) > 0:
                                #increase our normal counter
                                total+=1;
                                #increase our total normal counter
                                totalBooks+=1;
                                #add this file to our list of normal books
                                normalBooks.append(file)
                    #change our directory to one level above
                    os.chdir(os.path.dirname(os.getcwd()))
            #write the normal total to our file
            data_writer.writerow(["normal", total])
            #write the hate total to our file
            data_writer.writerow(["hate", hate])
            #write the adult total to our file
            data_writer.writerow(["adult",adult])
            #change the directory to one level above
            os.chdir(os.path.dirname(os.getcwd()))

#open a new csv file or overwrite the existing file and proceed to write in it
with open('hateAdultBookCount.csv', mode = 'w') as data:
    #initialize our writer object
    data_writer = csv.writer(data, delimiter=',', quoting=csv.QUOTE_MINIMAL)
    #write the header to our file
    data_writer.writerow(["book type","number"])
    #write the total amount of normal books to our file
    data_writer.writerow(["normal", totalBooks])
    #write the total amount of hate books to our file
    data_writer.writerow(["hate", totalHate])
    #write the total amount of adult books to our file
    data_writer.writerow(["adult",totalAdult])
#change our directory back to the intial directory
os.chdir(os.path.dirname(os.getcwd()))
#acquire the current path
cwd = os.getcwd()
#print statement signaling that the normal books list is about to be printed
print("These are Normal Books")
#iterate over the books in our normal books list
for book in normalBooks:
    #print every book in our list making sure to take out the ".csv" part of every book
    print( book.rsplit( ".", 1 )[ 0 ] )
```
